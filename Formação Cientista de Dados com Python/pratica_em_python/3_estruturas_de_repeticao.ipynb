```python
#exemplo com while
count=1
while count<=5:
    print(count)
    count += 1
```

```python
#range (0,10,1)
#range(inicio, parada, incremento)

for n in range(0,10):
    print(n+1)
for n in range(1,11):
    print(n)   
```

```python
#for decrescente
for n in range(10,0,-1):
    print(n)
```

```python
#exemplo com break
#ele para quando n==4
for n in range(10,0,-1):
    if n ==4:
        break
    print(n)
```

```python
#exemplo com continue
#ele nÃ£o apresenta o valor de n==4
for n in range(10,0,-1):
    if n ==4:
        continue
    print(n)
```
